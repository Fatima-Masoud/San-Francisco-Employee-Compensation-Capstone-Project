```python
import pandas as pd
import numpy as np

df = pd.read_csv("Employee_Compensation_Final.csv")

df.info()

print(df.describe().T)

invalid_summary = {
    "Negative_Values": (df[['Salaries','Overtime','Retirement','Health_and_Dental',
                            'Total_Benefits','Total_Compensation']] < 0).sum(),
    "Zero_Values": (df[['Salaries','Total_Compensation']] == 0).sum(),
    "Infinite_Ratios": np.isinf(df['Pay_to_Benefit_Ratio']).sum()
}
print(invalid_summary)
```

```python
df = df[
    (df['Overtime'] >= 0) &
    (df['Retirement'] >= 0) &
    (df['Health_and_Dental'] >= 0) &
    (df['Total_Benefits'] >= 0)
]
```

```python
df['Pay_to_Benefit_Ratio'] = df['Pay_to_Benefit_Ratio'].replace([np.inf, -np.inf], np.nan)
```

```python
df['Benefits_Percentage'] = np.clip(df['Benefits_Percentage'], 0, 100)
df['Overtime_Percentage'] = np.clip(df['Overtime_Percentage'], 0, 100)
```

```python
df['Union'] = df['Union'].fillna('Not Specified')
df['Employment_Type'] = df['Employment_Type'].replace('Nan', 'Not Specified').fillna('Not Specified')
```

```python
print(df.isna().sum())
print(df.describe())
```

```python
df.to_csv("Employee_Compensation_Accurate.csv", index=False)
```
