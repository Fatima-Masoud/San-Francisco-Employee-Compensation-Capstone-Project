```python
import pandas as pd
```

```python
df = pd.read_csv("Employee_Compensation_Cleaned.csv")
```

```python
df['Employment_Type'] = df['Employment_Type'].replace('Nan', 'Not Specified').fillna('Not Specified')
```

```python
df = df[(df['Salaries'] > 0) & (df['Total_Compensation'] < 1_000_000)]
```

```python
df['Overtime_Ratio'] = df['Overtime_Ratio'].fillna(0)
df[['Benefits_Percentage', 'Overtime_Percentage', 'Pay_to_Benefit_Ratio']] = \
    df[['Benefits_Percentage', 'Overtime_Percentage', 'Pay_to_Benefit_Ratio']].round(2)
```

```python
df.to_csv("Employee_Compensation_Final.csv", index=False)
```
